```python
analysisname = 'analysis_results-gpiemont'
typeofana = 'convolution'

path_with_analysis = Path(PATH_XE_OUTPUTS)

sufix = 'conv' if typeofana == "convolution" else 'deconv'
```

```python
def parse_module_key(mod):
    m = re.match(r'(C\d+)\((\d+)\)', mod)
    if m:
        return m.group(1), m.group(2)
    return mod, None
    
def load_module_images(path_to_data=f"{path_with_analysis}/{analysisname}/{typeofana}", sufix=sufix, run=None, modules=None):
    path_to_data = Path(path_to_data)
    images = {}
    
    if modules is None:
        png_files = list(path_to_data.rglob(f"convfit_plot_run0{run}*.png"))
                         
    elif modules == 'grid':
        png_files = list(path_to_data.rglob(f"convfit_grid_run0{run}*.png"))
        for file in png_files:
            img = mpimg.imread(file)
            plt.figure(figsize=(14, 10))
            plt.imshow(img)
            plt.axis("off")
            plt.tight_layout()
            plt.show()
        return {}
        
    else:
        png_files = []
        for mod in modules:
            mod_name, mod_idx = parse_module_key(mod)
            if mod_idx:
                png_files += list(path_to_data.rglob(f"*run0{run}*{mod_name}_{mod_idx}_*.png"))
            else:
                png_files += list(path_to_data.rglob(f"*run0{run}*{mod_name}_*.png"))
    
    for file in png_files:
        images[file] = mpimg.imread(file)
        
    return images
```
