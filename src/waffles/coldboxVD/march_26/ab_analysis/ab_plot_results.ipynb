```python
plot_config = {
    "snr_vs_vgain": {
        "x_me": "vgain",
        "y_me": "snr",
        "yerr_me": "e_snr",

        "x_fed": "VGain",
        "y_fed": "SNR",
        "yerr_fed": "Err SNR",

        "title": "SNR vs Vgain"
    },

    "snr_vs_spe": {
        "x_me": "spe_amplitude",
        "y_me": "snr",
        "yerr_me": "e_snr",

        "x_fed": "Spe ampl",
        "y_fed": "SNR",
        "yerr_fed": "Err SNR",

        "title": "SNR vs SPE"
    },

    "snr_vs_dr": {
        "x_me": "dynamic range",
        "y_me": "snr",
        "yerr_me": "e_snr",

        "x_fed": "DR",
        "y_fed": "SNR",
        "yerr_fed": "Err SNR",

        "title": "SNR vs Dynamic Range"
    },

    "spe_vs_vgain": {
        "x_me": "vgain",
        "y_me": "spe_amplitude",
        "yerr_me": "e_spe_amplitude",

        "x_fed": "VGain",
        "y_fed": "Spe ampl",
        "yerr_fed": "Err Spe ampl",

        "title": "SPE vs Vgain"
    }
}
```

```python
def plot_channel(type, module, channel,
                 vgain_scan_data_dict,
                 federico_results_dict,
                 bias_list,
                 plot_type,
                 comparison_federico,
                 plot_config,
                 save: bool = True,
                 output_dir: str = 'output',
                 additional_outputname : str = ''):

    cfg = plot_config[plot_type]

    cmap = plt.cm.tab10
    color_map = {bias: cmap(i % 10) for i, bias in enumerate(bias_list)}

    # ---- layout ----
    if comparison_federico:
        fig, axes = plt.subplots(1, len(bias_list), figsize=(5*len(bias_list), 5), sharey=True)
        axes = np.atleast_1d(axes)
    else:
        fig, ax = plt.subplots(figsize=(8,5))
        axes = [ax]

    for i, bias in enumerate(bias_list):

        ax = axes[i] if comparison_federico else axes[0]
        color = color_map[bias]

        # ======================
        # ME DATA
        # ======================
        filepath = vgain_scan_data_dict[type][module][channel][bias]
        df = pd.read_csv(filepath)

        ax.errorbar(
            df[cfg["x_me"]],
            df[cfg["y_me"]],
            yerr=df.get(cfg["yerr_me"], None),
            fmt='s',
            color=color,
            capsize=3,
            label=f"Bias {bias} DAC"
        )

        # ======================
        # FEDERICO
        # ======================
        if comparison_federico:
            fed_path = federico_results_dict[type][module][bias]
            df_fed = pd.read_csv(fed_path, index_col=False)
            df_fed["Err Spe ampl"] = 0

            df_fed = df_fed[
                (df_fed["Module"] == int(module.replace('M',''))) &
                (df_fed["DAPHNE Channel"] == int(channel)) &
                (df_fed["Bias [V]"] == int(bias))
            ]

            ax.errorbar(
                df_fed[cfg["x_fed"]],
                df_fed[cfg["y_fed"]],
                yerr=df_fed[cfg["yerr_fed"]],
                fmt='^',
                color=color,
                capsize=3,
                label="Federico"
            )

        ax.set_title(f"Bias {bias}")
        ax.set_xlabel(cfg["x_me"])
        ax.grid(True, alpha=0.3)
        ax.legend()

    axes[0].set_ylabel(cfg["y_me"])

    fig.suptitle(f"{type} module {module}, DAPHNE Channel {channel}\n{cfg['title']}")
    plt.tight_layout()

    if save:
        output_folder_analysis = f"{output_dir}/summary_results"
        if comparison_federico: 
            output_folder_analysis = f"{output_folder_analysis}/comparison_federico"
        if not os.path.exists(output_folder_analysis):
            os.makedirs(output_folder_analysis)
        plt.savefig(f"{output_folder_analysis}/{module}_ch{channel}_{plot_type}{additional_outputname}.png", dpi = 300)
    
    plt.show()
```
