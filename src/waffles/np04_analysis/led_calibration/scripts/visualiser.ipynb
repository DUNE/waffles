# Import required tools

```python
import os
import pickle
import datetime
import numpy as np
import pandas as pd
import plotly.graph_objects as pgo
import matplotlib.pyplot as plt
from waffles.np04_utils.utils import get_endpoint_and_channel
from waffles.np04_analysis.led_calibration.configs.calibration_batches.metadata import metadata
```

# Define some helper functions

```python
def compute_timestamp(day, month, year):
    """This function generates a timestamp from a date. 
    The date is given as three integer values which match 
    the day, the month and the year, respectively. The 
    timestamp is the number of seconds since 
    1970-01-01 00:00:00 UTC.
    
    """
    dt = datetime.datetime(year, month, day)
    return dt.timestamp()

def get_apa_foldername(measurements_batch, apa_no):
    """This function encapsulates the non-homogeneous 
    naming convention of the APA folders depending 
    on the measurements batch.""" 

    if measurements_batch not in [1, 2, 3]:
        raise ValueError(f"Measurements batch {measurements_batch} is not valid")
    
    if apa_no not in [1, 2, 3, 4]:
        raise ValueError(f"APA number {apa_no} is not valid")
                         
    if measurements_batch == 1:
        if apa_no in [1, 2]:
            return 'apas_12'
        else:
            return 'apas_34'
        
    if measurements_batch in [2, 3]:
        if apa_no == 1:
            return 'apa_1'
        elif apa_no == 2:
            return 'apa_2'
        else:
            return 'apas_34'
```

# Set the input variables

```python
apa_nos = [1, 2, 3, 4]
batches = [1, 2, 3, 4, 5, 6, 7]
variable = 'gain' # 'gain' or 'snr'
showlegend = False

input_csv_filepath = '../output/calibration_results.csv'
path_to_output_folderpath = '../output/'
batches_dates_csv_filepath = '../configs/batches_dates.csv'

if variable not in ['gain', 'snr']:
    raise Exception('Either gain or snr must be selected')
```

# Load the dataframe

```python
dataframe = pd.read_csv(input_csv_filepath)
dates = pd.read_csv(batches_dates_csv_filepath)
```

# Some input parameters for plots

```python
pdes = [0.50, 0.45, 0.40]

colors = {
    0.4: 'black', 
    0.45: 'green', 
    0.5: 'red'
}

symbols = {
    0.4: 'circle', 
    0.45: 'square',
    0.5: 'diamond'
}

translator = {
    'gain': 'Gain',
    'snr': 'SNR'
}

y_label = {
    'gain': 'center[1] - center[0]',
    'snr': '(center[1]-center[0])/sqrt( std[0]**2  + std[1]**2 )'
}

plot_y_minimum = {
    'gain': 1500,
    'snr': 1
}

plot_y_maximum = {
    'gain': 7500,
    'snr': 8
}
```

# Batch-wise plots

```python
# Arrange the time labels for the x-axis
time_labels = {}

for apa_no in apa_nos:
    for i in range(len(batches)):

        batch_no = batches[i]
        batch_date = dates[
            dates['batch'] == batch_no
        ]

        if len(batch_date) == 0:
            raise Exception(
                f"A date for batch {batch_no} was not "
                f"found in {batches_dates_csv_filepath}"
            )
        
        batch_day = batch_date['date_day'].values[0]
        batch_month = batch_date['date_month'].values[0]
        batch_year = batch_date['date_year'].values[0]
        
        time_labels[batch_no] = f"{batch_day}/{batch_month}/{batch_year}"
```

```python
for apa_no in apa_nos:

    for i in range(len(batches)):

        batch_no = batches[i]

        # Get the data for the given APA and batch
        current_df = dataframe[
            (dataframe['APA'] == apa_no) & 
            (dataframe['batch'] == batch_no)
        ]

        if len(current_df) == 0:
            print(
                f"WARNING: There is no data for APA {apa_no} and batch {batch_no}. "
                "Skipping this configuration..."
            )
            continue

        fig = pgo.Figure()

        for j in range(len(pdes)):

            aux = current_df[
                current_df['PDE'] == pdes[j]
            ]

            if len(aux) > 0:
                fig.add_trace(
                    pgo.Scatter(
                        x=aux['channel_iterator'],
                        y=aux[variable],
                        mode='markers',
                        marker=dict(
                            size=5, 
                            color=colors[pdes[j]],
                            symbol=symbols[pdes[j]]),
                        name=f"PDE = {pdes[j]}",
                    )
                )
        
        fig.update_yaxes(
            range=[
                plot_y_minimum[variable],
                plot_y_maximum[variable]
            ]
        )

        title = f"{translator[variable]} per channel in APA {apa_no} - "\
        f"Batch {batch_no} ({time_labels[batch_no]})"

        labels = {}
        for j in range(current_df.shape[0]):
            labels[current_df.iloc[j]['channel_iterator']] = \
                f"{int(current_df.iloc[j]['endpoint'])}-{int(current_df.iloc[j]['channel'])}"
        
        fig.update_layout(
            title={
                    'text': title,
                    'font': {'size': 18},
                },
            xaxis_title='Channel',
            yaxis_title=y_label[variable],
            width=1000,
            height=400,
            xaxis = dict(   
                tickmode='array',
                tickvals=list(labels.keys()),
                ticktext=list(labels.values()),
                tickangle=45,
            ),
            showlegend=True,
        )

        fig.show()
        fig.write_image(f"{path_to_output_folderpath}/batch_{batch_no}_apa_{apa_no}_pde_clustered_{variable}s.png")
```

```python
data = {}

for apa_no in apa_nos:
    data[apa_no] = {}

    for i in range(len(pdes)):
        data[apa_no][pdes[i]] = {}
        
        current_df = dataframe[
            (dataframe['APA'] == apa_no) &
            (dataframe['PDE'] == pdes[i])
        ]

        possible_channel_iterators = current_df['channel_iterator'].unique()
        
        for channel_iterator in possible_channel_iterators:
            data[apa_no][pdes[i]][channel_iterator] = {}

            aux = current_df[
                current_df['channel_iterator'] == channel_iterator
            ]

            x_time_ordered_values_of_variable = []
            y_time_ordered_values_of_variable = []

            # Here's why the data is ordered by batch number, i.e. ordered by time
            for batch_no in batches:

                aux2 = aux[
                    aux['batch'] == batch_no
                ]

                if len(aux2) == 0:
                    print(
                        f"WARNING: Found no entry for APA {apa_no}, PDE {pdes[i]}, "
                        f"batch {batch_no} and channel iterator {channel_iterator}."
                    )

                elif len(aux2) == 1:
                    x_time_ordered_values_of_variable.append(
                        batch_no
                    )
                    y_time_ordered_values_of_variable.append(
                        aux2[variable].values[0]
                    )

                else:
                    raise Exception(
                        f"Found more than one entry for APA {apa_no}, PDE {pdes[i]}, "
                        f"batch {batch_no} and channel iterator {channel_iterator}."
                    )

            data[apa_no][pdes[i]][channel_iterator]['batch_no'] = x_time_ordered_values_of_variable
            data[apa_no][pdes[i]][channel_iterator]['values'] = y_time_ordered_values_of_variable
```

# Std histograms, pde-wise

```python
bin_width = 40 if variable == 'gain' else 0.1

fig, axes = plt.subplots(
    nrows=1,
    ncols=len(apa_nos),
    figsize=(15, 5)
)

i = 0

for apa_no in data.keys():
    for pde in data[apa_no].keys():

        samples = []
        for channel_iterator in data[apa_no][pde].keys():

            if len(data[apa_no][pde][channel_iterator]['values']) == 0:
                continue

            else:
                samples.append(np.std(data[apa_no][pde][channel_iterator]['values']))

        samples = np.array(samples)

        axes[i].hist(
            samples, 
            bins=int((samples.max()-samples.min())/bin_width)+1,
            histtype="step", 
            color=colors[pde], 
            label=f"PDE = {pde}"
        )
        
        axes[i].set_xlabel(f"STD({y_label[variable]})")
        
        axes[i].legend(
            fontsize = 15
        )

        axes[i].grid()
        
    axes[i].set_title(f"APA {apa_no}")
    i += 1
            
title = f"{translator[variable]} STD (over time)"

fig.suptitle(title, fontsize=16)
plt.tight_layout()
plt.show()
```
